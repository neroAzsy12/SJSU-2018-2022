```python
covid_aln = AlignIO.read(open('Phylogenetic Trees/covid_tree_construction.txt'), 'phylip')
covid_constructor = DistanceTreeConstructor()
covid_calculator = DistanceCalculator('blosum62')

covid_dm = covid_calculator.get_distance(covid_aln)
covid_upgmatree = covid_constructor.upgma(covid_dm) 
```

```python
Phylo.write(covid_upgmatree, 'Phylogenetic Trees/covid_tree.xml', 'phyloxml') 
```
