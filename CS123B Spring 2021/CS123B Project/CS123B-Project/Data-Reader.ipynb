```python
import numpy as np
import pandas as pd
import os 
```

```python
np.random.seed(42) 
```

```python
class Sequence: # 4
    def __init__(self, seq_name, seq_data):
        self.name = seq_name
        self.data = seq_data
```

```python
covid_data = [] #17

covid_file = open('Sequences Data/Combined Data/unaligned_covid_data.fasta', 'w')

for filename in os.listdir('Sequences Data/Coronavirus Data'):
    nameOfFile = os.path.join('Sequences Data/Coronavirus Data', filename)
    file = open(nameOfFile, 'r')

    sequence_name = file.readline()
    
    sequence_data = ''
    lines = file.read().splitlines()
    
    for line in lines:
           sequence_data += line
    
    covid_data.append(Sequence(sequence_name[1 : len(sequence_name) - 1], sequence_data))
    
    covid_file.write(sequence_name)
    covid_file.write(sequence_data + '\n\n')
    
    file.close()

covid_file.close()

for data in covid_data:
    print(data.name)
```

```python
non_covid_data = [] #17
non_covid_file = open('Sequences Data/Combined Data/unaligned_non_covid_data.fasta', 'w')

for filename in os.listdir('Sequences Data/Non-Covid Data'):
    nameOfFile = os.path.join('Sequences Data/Non-Covid Data', filename)
    file = open(nameOfFile, 'r')
    
    sequence_name = file.readline()
    
    sequence_data = ''
    lines = file.read().splitlines()
    
    for line in lines:
           sequence_data += line
    
    non_covid_data.append(Sequence(sequence_name[1 : len(sequence_name) - 1], sequence_data))
    
    non_covid_file.write(sequence_name)
    non_covid_file.write(sequence_data + '\n\n')
    
    file.close()

non_covid_file.close()

for data in non_covid_data:
    print(data.name)
```

## Accessing the fasta files that were obtained from the PHMMs

### PHMM for Non-Covid

```python
class SequenceScore:
    def __init__(self, label, name, length, score): #6
        self.label = label
        self.name = name
        self.length = length
        self.score = score
```

```python
phmm_noncovid = [] #38

with open('PHMM Files/PHMM Noncovid/noncovid_training_results', 'r') as file:
    lines = file.read().splitlines()
    
    score, name = lines[0][19:].split('__')
    sequence = ''
    
    for i in range(1, len(lines)):
        if lines[i][0] == '>':
            phmm_noncovid.append(SequenceScore('Non-Covid Training', name, len(sequence.replace('-', '')), float(score)))
            score, name = lines[i][19:].split('__')
            sequence = ''
            
        else:
            sequence += lines[i]
    
    phmm_noncovid.append(SequenceScore('Non-Covid Training', name, len(sequence.replace('-', '')), float(score)))

with open('PHMM Files/PHMM Noncovid/noncovid_testing_results', 'r') as file:
    lines = file.read().splitlines()
    
    score, name = lines[0][19:].split('__')
    sequence = ''
    
    for i in range(1, len(lines)):
        if lines[i][0] == '>':
            phmm_noncovid.append(SequenceScore('Non-Covid Testing', name, len(sequence), float(score)))
            score, name = lines[i][19:].split('__')
            sequence = ''
            
        else:
            sequence += lines[i]
    
    phmm_noncovid.append(SequenceScore('Non-Covid Testing', name, len(sequence), float(score)))

with open('PHMM Files/PHMM Noncovid/covid_testing_results', 'r') as file:
    lines = file.read().splitlines()
    
    score, name = lines[0][19:].split('__')
    sequence = ''
    
    for i in range(1, len(lines)):
        if lines[i][0] == '>':
            phmm_noncovid.append(SequenceScore('Covid Testing', name, len(sequence), float(score)))
            score, name = lines[i][19:].split('__')
            sequence = ''
            
        else:
            sequence += lines[i]
    
    phmm_noncovid.append(SequenceScore('Covid Testing', name, len(sequence), float(score)))

print(len(phmm_noncovid))
```

## Data frame to properly show the data from the PHMM

```python
phmm_noncovid.sort(key = lambda x: x.score, reverse = True) #1
```

```python
data = [] #6

for sequence in phmm_noncovid:
    data.append([sequence.name, sequence.score, sequence.length, sequence.label])
    
noncovid_df = pd.DataFrame(data, columns = ['Sequence Name', 'Log Viterbi Score', 'Amino Acid Length', 'Data Origin'])
noncovid_df.index += 1

noncovid_df
```

```python
noncovid_df.to_csv('Non-Covid PHMM Results.csv') #1
```
